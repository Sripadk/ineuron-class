```python
# To find the relation between sepal legth and sepal width
df.plot.scatter(x = 'sepal_length', y='sepal_width')
# This shows that there is no excat relateion 
# The data set is excatly distributed
```

```python
# To find the relation between sepal legth and petal length
df.plot.scatter(x = 'sepal_length', y='petal_length')
# This shows that as sepal length increases the petal length also increases.
# thus there is a linear relation between sepal length and petal length.
```

```python
# Colour control 
df.plot.scatter(x = 'sepal_length', y='petal_length', c = 'sepal_width')

# c = 'sepal_width' : This will show the bar for sepal_width.
# sepal width starts from 2 and will go till 4.5
# as the scale progressess the bar becomes darker.
# The colour of each scatter plot dot is decided by the bar of sepal_width.
# lighter colour shows smaller alue of sepal _width at that dot.
# darker colour of the dot shows that value of sepal_width is high at that dot.
```

```python
df.plot.scatter(x = 'sepal_length', y='petal_length', c = 'sepal_width', s = 10)
# s= 10: this is the size of the dot scatter plot
# s = 10 to 1000. 
# s = df['sepal_length']*20
```

```python
df.plot.scatter(x = 'sepal_length', y='petal_length', s = 40)
# s = df['sepal_length']*40 : this varies the size depending on the column values 'sepal_lenght'
```

## Please use online matplot library or seaborn or plotly library to get each type of graphs.

```python
# joint plot. This provides two plots w.r.t the same data set as below.
# graphs are scatter plot and histogrram.

sns.jointplot(x = df1.total_bill, y = df1.tip)
```

```python
# instead of scatter plot if we want hexabin plot. then change the kind

sns.jointplot(x = df1.total_bill, y = df1.tip, kind = 'hex')
```

```python
# pairplot
sns.pairplot(df1)
```

```python
# Regression plot.
sns.regplot(x = df1.total_bill, y = df1.tip)
# This will also provide the best fit regression line.
# Now to change the fig size in seaborn you need to execute the below line
sns.set(rc = {'figure.figsize':(20,10)})
```
